## Part 1: Camera Calibration

Undistort checkerboard images 

### Briefly state how you computed the camera matrix and distortion coefficients. Provide an example of a distortion corrected calibration image.

You can find my code for the camera calibration in the `src/camera_calibrator.py` file. 
- First find the object points for the images. Because our calibration images contain the same object points, this can simply be reused for every image in the calibration set.
- Convert each image to greyscale and use `cv2.findChessboardCorners`. This will give you the corners as well as a boolean that indicates whether the corners were successfully found. If they were found, I use `cv2.drawChessboardCorners` to visualize the corners and save the corners and object points for that image.
- Once I have the object points and image points (corners), I use those values to calibrate the camera using `cv2.calibrateCamera`. This gives me the camera matrix and distortion coefficients, that I can pass into the `cv2.undistort` method in order to undistort the images.

Examples of each step are listed below:

```python
from camera_calibrator import CameraCalibrator, CORNERS_PATH, ORIGINAL_PATH, UNDISTORTED_PATH, WARPED_PATH
CALIBRATION_PATH = '../output_images/calibrations.p'
SAMPLE_INDEX = 2

camera = CameraCalibrator(9, 6)
camera.distortion_correction('../camera_cal/calibration*.jpg')
cam_saver = camera.output_saver
cam_saver.on_end(CALIBRATION_PATH)
```

### Step 1: Detect Corners 

```python
org_image = cam_saver.images[ORIGINAL_PATH][SAMPLE_INDEX]
new_image = cam_saver.images[CORNERS_PATH][SAMPLE_INDEX]
cam_saver.plot_images('Original Image', org_image, 'Image with Detected Corners', new_image)
```

### Step 2: Undistort Images

```python
org_image = cam_saver.images[ORIGINAL_PATH][SAMPLE_INDEX]
new_image = cam_saver.images[UNDISTORTED_PATH][SAMPLE_INDEX]
cam_saver.plot_images('Original Image', org_image, 'Undistorted Image', new_image)
```

### Step 3: Perspective Transform

```python
org_image = cam_saver.images[ORIGINAL_PATH][SAMPLE_INDEX]
new_image = cam_saver.images[WARPED_PATH][SAMPLE_INDEX]
cam_saver.plot_images('Original Image', org_image, 'Image with Perspective Transform', new_image)
```
